# 1. Bagging

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
%matplotlib inline
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC      
from sklearn.tree import DecisionTreeClassifier

from sklearn.naive_bayes import GaussianNB
```

```python
iris= datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)  # datasets 下來格式非DF,需轉換
iris_df['target'] = iris.target
```

```python
iris_df.columns
```

```python
y = iris_df['target']
X = iris_df.drop(columns = 'target')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =0)
```

```python
vc = VotingClassifier(estimators = [('svm',SVC()),('dc',DecisionTreeClassifier()),('nb',GaussianNB())])
```

```python
vc
```

```python
vc.fit(X_train,y_train)
y_pred = vc.predict(X_test)
```

```python
vc.score(X_test,y_test)
```

# 2. Boosting

```python
from sklearn.ensemble import AdaBoostClassifier  # 回歸問題 : AdaBoostRegressor
from sklearn.ensemble import GradientBoostingClassifier  # 回歸問題 : GradientBoostingRegressor
```

```python
adb= AdaBoostClassifier()
adb.fit(X_train,y_train)
adb.predict(X_test)
adb.score(X_test,y_test)
```

```python
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
gb.predict(X_test)
gb.score(X_test,y_test)
```
